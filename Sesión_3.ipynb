# Algebra lineal

$
\huge
\mathbf{A}\cdot\ x=\vec{b}\ \therefore\ x=\mathbf{A^{-1}}\cdot\vec{b}
$

$
\huge
\mathbf{A}
$
$=\begin{pmatrix}
a_{1,1} & \dots & a_{1,n}\\
\vdots & \ddots & \vdots \\
a_{1,1} & \dots & a_{m,n}
\end{pmatrix}
$

$
\huge
x=
$
$
\begin{pmatrix}
x_1\\
\vdots\\
x_m
\end{pmatrix}
$

$
\huge
\vec{b}=
$
$
\begin{pmatrix}
b_1\\
\vdots\\
b_m
\end{pmatrix}
$

$
\large
\begin{matrix}
x-3y+2z=-3\\
5x+6y-z=13\\
4x-y+3z=8
\end{matrix}
$

```
import numpy as np
import matplotlib.pyplot as plt
```

```
A=np.array([[1,-3,2],
            [5,6,-1],
            [4,-1,3]])
A
```

```
b=np.array([-3,13,8])
b
```

$
\large
\mathbf{A^{-1}}=\frac{adj\mathbf{(A^T)}}{\det{\mathbf{A}}}
$

```
def minor(matriz,i,j):
  i-=1
  j-=1
  import numpy as np
  matriz=list(matriz)
  x=[list(fila) for fila in matriz[:i]+matriz[i+1:]]
  minor=[fila[:j]+fila[j+1:] for fila in x]
  return np.array(minor)
```

# Matriz de rotación

```
theta=np.radians(0)

rot_z=np.array([[np.cos(theta),-np.sin(theta)],
                [np.sin(theta),np.cos(theta)]])
```

```
rot_z
```

```
def rotacion_z(grados):
  theta=np.radians(grados)
  rot_z=np.array([[np.cos(theta),-np.sin(theta)],
                  [np.sin(theta),np.cos(theta)]])
  return rot_z
```

```
rotacion_z(75)
```

```
a=np.array([5,3])
```

```
plt.arrow(0,0,a[0],a[1], color="g", head_width=0.75, head_length=0.75)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid()
plt.show()
```

```
rotacion_z(45).dot(a)
```

```
fig, axs= plt.subplots(figsize=(5,5))
axs.arrow(0,0,a[0],a[1], color="g", head_width=0.75, head_length=0.75)
axs.text(a[0]*1.2,a[1]*1.2,"a", size=14)
axs.arrow(0,0,rotacion_z(90).dot(a)[0],rotacion_z(90).dot(a)[1], color="r", head_width=0.75, head_length=0.75)
axs.text(rotacion_z(90).dot(a)[0]*1.2,rotacion_z(90).dot(a)[1]*1.2,"a'", size=14)
axs.set_xlim(-10,10)
axs.set_ylim(-10,10)
axs.grid()
plt.show()
```

```
def plano_xy(vector, angulo, ax=axs):
  axs.arrow(0,0,vector[0],vector[1], color="g", head_width=0.75, head_length=0.75)
  axs.text(vector[0]*1.2,vector[1]*1.2,"a", size=14)
  axs.arrow(0,0,rotacion_z(angulo).dot(vector)[0],rotacion_z(angulo).dot(vector)[1], color="r", head_width=0.75, head_length=0.75)
  axs.text(rotacion_z(angulo).dot(vector)[0]*1.2,rotacion_z(angulo).dot(vector)[1]*1.2,"a'", size=14)
  maximo=max(abs(vector[0]),abs(vector[1]))
  axs.set_xlim(-maximo*2,maximo*2)
  axs.set_ylim(-maximo*2,maximo*2)
  axs.grid()
```

```
fig, axs= plt.subplots(figsize=(5,5))
plano_xy(np.array([-3,-15]),125)
plt.show()
```
